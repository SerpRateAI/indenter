## Importing Libraries

```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
from PIL import Image
import math
from pathlib import Path
from merrypopins.make_dataset import merrypopins_pipeline
```

## Define the Datasets Folder and the Output Folder for Visualizations

```python
# Directory containing all the txt files
data_dir = Path("datasets/6microntip_slowloading")
output_dir = Path("visualisations/6microntip_slowloading")
```

```python
# Make sure output directory exists
output_dir.mkdir(parents=True, exist_ok=True)
```

## Run the MerryPopins Pipeline

```python
# Loop through all txt files in the dataset directory
for txt_file in data_dir.glob("*.txt"):
    print(f"Processing {txt_file.name}...")
    try:
        df_loc = merrypopins_pipeline(
            txt_path=txt_file,
            save_plot_dir=output_dir,
            trim_margin=30
        )
        print(f"✅ Done: {txt_file.name} (Pop-ins detected: {df_loc['popin'].sum()})")
    except Exception as e:
        print(f"❌ Failed: {txt_file.name} with error: {e}")
```

## Show the Visualizations

```python
# Load all PNGs from output folder
image_paths = sorted(output_dir.glob("*.png"))

# Only proceed if there are images
if image_paths:
    n_images = len(image_paths)
    print(f"Found {n_images} plots in output folder.")
    n_cols = 3 # Number of columns for the plot grid
    n_rows = math.ceil(n_images / n_cols)

    fig, axs = plt.subplots(n_rows, n_cols, figsize=(12, 6 * n_rows))

    for ax, img_path in zip(axs.flatten(), image_paths):
        img = Image.open(img_path)
        ax.imshow(img)
        ax.set_title(img_path.stem)
        ax.axis('off')

    # Hide any unused subplots
    for ax in axs.flatten()[n_images:]:
        ax.axis('off')

    plt.tight_layout()
    plt.show()
else:
    print("No plots found in output folder.")
```
